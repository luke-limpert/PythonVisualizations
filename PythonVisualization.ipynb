```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import requests
import io
```

```python
#import data from website
url = 'https://www.metoffice.gov.uk/pub/data/weather/uk/climate/stationdata/heathrowdata.txt'
file = io.StringIO(requests.get(url).text)
```

```python
#name Columns
col_names = ('Year','Month','Tmax','Tmin','AF','Rain','Sun')
comment_lines = 4
header = 2
```

```python
#create data frame from .csv
weather = pd.read_csv(file,
   skiprows=comment_lines + header,
   header=0,
   names=col_names,
   delimiter=' ',
   skipinitialspace=True,
   error_bad_lines=False)
```

```python
#render data frame
weather
```

```python
file.seek(0)

col_names = ('Year','Month','Tmax','Tmin','AF','Rain','Sun', 'Status')
rows_to_skip = comment_lines+header+len(weather)

weather2 = pd.read_csv(file,
   skiprows=rows_to_skip,
   header=0,
   names=col_names,
   delimiter=' ',
   skipinitialspace=True,
   error_bad_lines=False)
```

So, on In [4]: We get an error that 7 rows were expected, but 8 were seen. This still made the data frame because of the "error_bad_lines = False". 

To correct, we need to send the cursor back to the beginning of the file to read it again. This is what seek(0) does. We also add the len(weather) to skip everything we have done before, which will make the program faster. 

```python
weather2
```

Updated data frame with the status column

In order to merge the 2 data frames they need to be the same shape. So, you have the choice of deleting the status column, or adding a status column to the weather data frame. We went through the effort of adding the column for the 2020 data in weather2 dataframe, so we will add a column to the weather dataframe. 

```python
weather['Status']='Final'
```

```python
weather = weather.append(weather2, ignore_index=True)
#ignore_index creates a new index for the combined dataframe instead of duplicating the indices
```

```python
weather
```

2020 is now apart of the data set and the status has been appended to all of the records from the original weather dataframe

```python
weather.dtypes
```

As you can see in the rendering above, the ['Sun'] columm is numeric, but has a # suffix which is causing it to be perceived as an object. 

```python
weather['Sun']=weather['Sun'].str.replace('#','')
# remove the #
```

```python
weather['AF']=pd.to_numeric(weather['AF'], errors='coerce')
weather['Sun']=pd.to_numeric(weather['Sun'], errors='coerce')

#We set the errors to coerce so that any values that cannot be interpreted as numbers will be set to NaN.
```
